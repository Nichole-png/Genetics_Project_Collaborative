# Cholesterol Homeostasis: Low Density Lipoprotein Receptor



## Introduction

Cholesterol and triacylglycerols are hydrophobic molecules, so require a special transporting system through the blood.  They are carried in lipoprotein particles which are composed of a hydrophobic core and hydrophilic exterior.  Low density lipoprotein (LDL) contains apoprotein B-100 (Apo B-100) at its surface.  Apo B-100 interacts with the low density lipoprotein receptor and leads to receptor-mediated endocytosis.  If any part of this system is not functioning optimally, plasma cholesterol can increase above normal levels.  This increase in plasma cholesterol (or LDL) is a primary risk factor for atherosclerosis.

One of the sources of increased LDL is a reduced efficiency of the LDL receptor which is found on the cells on the exterior of the liver and intestine.  The LDL receptor (LDLR) is a transmembrane protein with the majority of the protein on the extracellular side of the membrane, ready to interact with LDL in the plasma.
